```python
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import requests
import io
```

```python
#downloading the image
url="https://www.sciencenewsforstudents.org/sites/default/files/2016/12/main/articles/860_main_beauty.png"
image=Image.open(io.BytesIO(requests.get(url,stream=True).content)).convert("RGB")
```

```python
plt.imshow(image)
```

```python
image_array=np.array(image)
```

```python
gray=cv.cvtColor(image_array,cv.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
detector=cv.CascadeClassifier("C:\\OpenCV\\opencv\\sources\\data\\haarcascades\\haarcascade_frontalface_default.xml")
```

```python
faces=detector.detectMultiScale(gray,1.3,5)
```

```python
faces[0]
```

```python
x,y,w,h=faces[0]
```

```python
cv.rectangle(image_array,(x,y),(x+w,y+h),(255,255,255),2)
plt.imshow(image_array)
```
