## Image Warping - Waves

```python
import math
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
img = cv2.imread('images\\dp.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
rows, cols = img.shape
```

```python
rows, cols
```

```python
img[:2].dtype
```

```python
img_output = np.zeros(img.shape[:2])
```

```python
img
```

```python
for i in range(rows):
    for j in range(cols):
        
        offset_x = int(105.0*math.sin(2*3.14*i / 180))
        offset_y = 0
        
        if j+offset_x < cols:
            img_output[i,j] = img[i,(j+offset_x)]
        else:
            img_output[i,j] = 0
```

```python
cv2.imwrite('Wave_Y.png',img_output)
```

```python
for i in range(rows):
    for j in range(cols):
        
        offset_x = 0
        offset_y = int(55.0*math.sin(2*3.14*j / 180))
        
        if i+offset_y < rows:
            img_output[i,j] = img[(i+offset_y),j]
        else:
            img_output[i,j] = 0
```

```python
cv2.imwrite('Wave_X.png',img_output)
```

## Image Blurring

```python
img = cv2.imread('images\\dp.jpg')
```

```python
rows, cols = img.shape[:2]
```

```python
kernel_idn = np.array([[0,0,0], [0,1,0], [0,0,0]])
```

```python
kernel_3 = np.ones((3,3), dtype=np.float32) / 9.0
```

```python
kernel_5 = np.ones((5,5), dtype=np.float32) / 25.0
```

```python
output3 = cv2.filter2D(img, -1, kernel_3)
```

```python
output5 = cv2.filter2D(img, -1, kernel_5)
```

```python
plt.imshow(img)
```
